```python
import tensorflow as tf
from tensorflow import keras
import os, datetime
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
def create_model():
  model = tf.keras.models.Sequential([
      tf.keras.layers.Flatten(input_shape=(28, 28)),
      tf.keras.layers.Dense(128, activation = 'selu'),
      tf.keras.layers.Dense(10, activation='softmax')
  ])

  model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
  return model
model = create_model()
```

```python
check1 = "training1/cp.ckpt"
check_dir1 = os.path.dirname(check1)
```

```python
cb = tf.keras.callbacks.ModelCheckpoint(
    check1, verbose=1, save_weights_only = True,
    save_freq = 'epoch'
)

model.fit(x_train, y_train, epochs = 10, validation_data = (x_test, y_test), callbacks = [cb])
```

```python
model1 = create_model()
model1.load_weights(check1)
loss, acc = model1.evaluate(x_test, y_test)
```

```python
check2 = "training2/cp-{epoch:04d}.ckpt"
check2_dir = os.path.dirname(check2)

cb2 = tf.keras.callbacks.ModelCheckpoint(
    check2, verbose=1, save_weights_only=True,
    save_freq = 'epoch'
)
model = create_model()
model.fit(x_train, y_train, epochs = 10,
          validation_data = (x_test, y_test), callbacks=[cb2])
```

```python
model2 = create_model()
most_recent = tf.train.latest_checkpoint(check2_dir)

model2.load_weights(most_recent)
loss, acc = model.evaluate(x_test, y_test)
```

```python
check3 = "training3/cp-{epoch:04d}-accuracy-{accuracy:.2f}.ckpt"
check3_dir = os.path.dirname(check3)

cb3 = tf.keras.callbacks.ModelCheckpoint(
    check3, verbose=1, save_weights_only=True,
    save_freq = 'epoch'
)

model =create_model()

model.fit(x_train, y_train, epochs = 10,
          validation_data = (x_test, y_test), callbacks=[cb3])
```

```python
check4 = "training_4/ckpt-loss={loss:.4f}"
check4_dir = os.path.dirname(check4)

cb4 = tf.keras.callbacks.ModelCheckpoint(
   filepath=check4, save_weights_only=True, verbose = 1,
   save_freq=100)
model = create_model()

model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), callbacks=[cb4])
```

```python
check5 = "training_5/ckpt-{epoch:04d}-acc={accuracy:.3f}"
check5_dir = os.path.dirname(check5)

cb5 = tf.keras.callbacks.ModelCheckpoint(
   filepath=check5, save_weights_only=True, verbose = 1,
   monitor = 'accuracy', mode='auto',
   save_freq=100, save_best_only = True)
model = create_model()

model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), callbacks=[cb5])
```
